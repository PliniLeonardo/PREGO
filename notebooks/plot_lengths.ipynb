```python
import pickle
from matplotlib import pyplot as plt
```

```python
file_path = "/media/hdd/usr/edo/egoProcel_mistakes/results/out_plot_1_assembly.pkl"
data = pickle.load(open(file_path, "rb"))
```

```python
lens_hist, lens_pred = [], []
for k, v in data.items():
    lens_hist.append(k)
    lens_pred.append(v["sum"] / v["count"])
```

```python
# plot lens_hist on the x axis and lens_pred on the y axis
plt.plot(lens_hist, lens_pred)
# write the labels for the x and y axis
plt.xlabel("Input sequence length")
plt.ylabel("Average unique predicted choices")
```
