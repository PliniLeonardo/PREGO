```python
from pathlib import Path
import json
```

```python
classification_json = (
    "/media/hdd/usr/edo/egoProcel_mistakes/results/classification.json"
)
classification_json = json.load(open(classification_json, "r"))
```

```python
preds = classification_json["pred"]
gts = classification_json["GT"]
idxs = classification_json["session_indexes"]
names = classification_json["session_names"]
```

```python
from requests import session


out = {}
for i in range(len(names)):
    if names[i] in out:
        print("error")
    s, e = idxs[i], idxs[i + 1]
    out[names[i]] = {"pred": preds[s:e], "gt": gts[s:e]}


print(out)
```

```python
# Save to json
out_path = Path("/media/hdd/usr/edo/egoProcel_mistakes/results/classification_.json")
json.dump(out, open(out_path, "w"))
```
